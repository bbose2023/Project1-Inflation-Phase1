```python
##  IN STATE TUITION INFLATION
```

```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Define the file path
file_path = Path("Resources/instate_tuition.csv")
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
print(df.head())
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv(file_path, header=1)  # Use the second row as header

# Remove any leading/trailing spaces from column names
df.columns = df.columns.str.strip()

# Rename columns 'Unnamed: 0' and 'Unnamed: 1'
df.rename(columns={'Unnamed: 0': 'Institution', 'Unnamed: 1': 'STATE'}, inplace=True)

# Display the first few rows to understand the structure
print(df.head())
```

```python
# Melt the DataFrame to long format
df_melted = pd.melt(df, id_vars=["Institution", "STATE"], var_name="Year", value_name="Tuition")

# Extract only relevant rows (e.g., years and tuition data)
# You may need to filter out non-year rows (like '1-Year' and '5-Year' percentage change)
df_melted = df_melted[df_melted["Year"].str.contains(r'^\d{4}-\d{2}$')]

# Optionally, rename columns for clarity
df_melted.rename(columns={"STATE": "State", "Tuition": "Tuition (USD)"}, inplace=True)

# Convert Tuition to numeric, removing any unwanted characters like commas or dollar signs
df_melted["Tuition (USD)"] = df_melted["Tuition (USD)"].replace('[\$,]', '', regex=True).astype(float)

# Drop the 'Institution' column
df_melted_no_institution = df_melted.drop(columns=["Institution"])

print(df_melted_no_institution)
```

```python
# Calculate the average tuition across all years for each state
average_tuition = df_melted_no_institution.groupby('State')['Tuition (USD)'].mean().sort_values(ascending=False).reset_index()

# Merge this average tuition back to the original data
df_sorted = pd.merge(df_melted_no_institution, average_tuition, on='State', suffixes=('', '_Average'))

# Sort the data based on the average tuition
df_sorted = df_sorted.sort_values(by='Tuition (USD)_Average', ascending=False)

print(df_sorted)
```

```python
# Plot the bar plot for all years, with states sorted by average tuition
plt.figure(figsize=(16, 10))
ax=sns.barplot(data=df_sorted, x="State", y="Tuition (USD)", hue="State", palette="viridis")
plt.title("Comparison of Average In-State Tuition Fees for Years(2013-2023)", fontsize=17)
plt.xticks(rotation=90, fontsize=12)
plt.xlabel("State", fontsize=15)
plt.ylabel("Tuition (USD)", fontsize=15)
plt.yticks(fontsize=12)
ax.yaxis.set_major_locator(plt.MaxNLocator(10))
plt.grid(True)
plt.tight_layout()

#Save the plot as an image in the specified folder
folder_name = "Images"
image_path = Path(folder_name, "Ed-instate_tuition_comparison_plot.png")
plt.savefig(image_path)

plt.show()
```

```python
# Calculate the average tuition across all states for each year
yearly_avg_tuition = df_melted_no_institution.groupby('Year')['Tuition (USD)'].mean().reset_index()

# Rename the column for clarity
yearly_avg_tuition.rename(columns={"Tuition (USD)": "Yearly Average Tuition (USD)"}, inplace=True)

# Merge this yearly average tuition back to the original data
df_with_yearly_avg = pd.merge(df_melted_no_institution, yearly_avg_tuition, on="Year")

# Display the updated DataFrame
print(df_with_yearly_avg)
```

```python
# Calculate year-over-year growth rates for each state
df_growth = df_melted_no_institution.copy()
df_growth['Tuition Growth (%)'] = df_growth.groupby('State')['Tuition (USD)'].pct_change() * 100

# Drop the rows where 'Year' is '2013-14' and 'Tuition Growth (%)' is NaN
df_growth = df_growth[~((df_growth['Year'] == '2013-14') & (df_growth['Tuition Growth (%)'].isna()))]

# Pivot the table to create a heatmap
pivot_growth = df_growth.pivot(index="State", columns="Year", values="Tuition Growth (%)")
```

```python
# Plot the heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(pivot_growth, cmap="viridis", annot=True, fmt=".2f", linewidths=.5, center=0)
plt.title("Heatmap of Percent Change in In-State Tuition Cost", fontsize=15)
plt.xticks(rotation=45)
plt.xlabel("Academic Year", fontsize=15)
plt.ylabel("State", fontsize=15)
plt.yticks(rotation=0)
plt.tight_layout()

#Save the plot as an image in the specified folder
folder_name = "Images"
image_path = Path(folder_name, "Ed-instate_tuition_heatmap.png")
plt.savefig(image_path)

plt.show()
```

```python
# Calculate the average growth rate for each year across all states
average_growth_per_year = df_growth.groupby('Year')['Tuition Growth (%)'].mean().reset_index()
```

```python
# Calculate the average growth rate for in-state tuition
average_growth_instate = df_growth.groupby('Year')['Tuition Growth (%)'].mean().reset_index()
average_growth_instate['Type'] = 'In-State'

# Calculate the average growth rate for out-state tuition
average_growth_outstate = df_out_growth.groupby('Year')['Tuition Growth (%)'].mean().reset_index()
average_growth_outstate['Type'] = 'Out-State'

# Combine the two DataFrames
combined_growth = pd.concat([average_growth_instate, average_growth_outstate])

# Set the style for the plot
sns.set(style="darkgrid")

# Plot the combined line graph for average tuition growth percentage over the years
plt.figure(figsize=(12, 8))
ax = sns.lineplot(data=combined_growth, x="Year", y="Tuition Growth (%)", hue="Type", marker="o",)

# Annotate the values on the plot
for i in range(combined_growth.shape[0]):
    year = combined_growth['Year'].iloc[i]
    growth = combined_growth['Tuition Growth (%)'].iloc[i]
    tuition_type = combined_growth['Type'].iloc[i]
    ax.text(year, growth, f'{growth:.2f}', color='red', ha="center", va="bottom")

# Customize the plot
plt.title('Comparison of Average Change in Tuition : In-State vs Out-State (2013-2023)', fontsize=16)
plt.xlabel('Academic Year', fontsize=14)
plt.ylabel('Change (%)', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Tuition Type', fontsize=12)
plt.grid(True)

#Save the plot as an image in the specified folder
folder_name = "Images"
image_path = Path(folder_name, "Ed-comparision_invsout.png")
plt.savefig(image_path)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import requests
from pathlib import Path
import pandas as pd
import json
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
```

```python
# Define the API endpoint URL
url_1 = "https://api.stlouisfed.org/fred/series/observations?series_id=CUSR0000SAE1&api_key=2d0f11a14423dbcf9beaabf0add69972&file_type=json"
response = requests.get(url_1)
data = response.json()

# Pretty print the JSON data
print(json.dumps(data, indent=4))
```

```python
# Extract the observations
observations = data['observations']

# Extract dates and values
dates = [obs['date'] for obs in observations]
values = [obs['value'] for obs in observations]

# Print the first few entries to verify
print("Dates:", dates[:5])
print("Values:", values[:5])
```

```python
# Convert to DataFrame
df = pd.DataFrame(observations)

# Convert 'date' to datetime format and 'value' to numeric
df['date'] = pd.to_datetime(df['date'])
df['value'] = pd.to_numeric(df['value'])

# Display the first few rows of the DataFrame
print(df.head())
```

```python
# Filter data between 2013 and 2023
start_date = '2013-01-01'
end_date = '2023-12-31'
filtered_df = df[(df['date'] >= start_date) & (df['date'] <= end_date)]

# Plot the filtered data as a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(filtered_df['date'], filtered_df['value'], color='darkred')

# Set x-axis to show only years
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator())

plt.xlabel('Year')
plt.ylabel('Index Value')
plt.title('CPI for All Urban Consumers: Education in U.S. City Average')
plt.grid(True)

#Save the plot as an image in the specified folder
folder_name = "Images"
image_path = Path(folder_name, "Ed-CPI_USavg.png")
plt.savefig(image_path)

plt.show()
```

```python
import requests
import pandas as pd
from pathlib import Path
import json
import matplotlib.pyplot as plt
```

```python
# Define the API endpoint URL
url_2 = "https://api.stlouisfed.org/fred/series/observations?series_id=CXUFEESADMLB1408M&api_key=2d0f11a14423dbcf9beaabf0add69972&file_type=json"
response_2 = requests.get(url_2)
data_2 = response_2.json()

# Pretty print the JSON data
print(json.dumps(data_2, indent=4))
```

```python
# Convert observations to DataFrame
df_2 = pd.DataFrame(data_2['observations'])

# Convert 'date' to datetime format and 'value' to numeric
df_2['date'] = pd.to_datetime(df_2['date'])
df_2['value'] = pd.to_numeric(df_2['value'])

# Display the first few rows of the DataFrame
print(df_2)
```

```python
# Extract the year from the date and create a new column
df_2['year'] = df_2['date'].dt.year

# Calculate Year-over-Year percentage change
df_2['Inflation Rate(%)'] = df_2['value'].pct_change() * 100

# Drop NaN values from the Inflation Rate(%) column
df_2 = df_2.dropna(subset=['Inflation Rate(%)'])

print(df_2)
```

```python
# Plot the Year-over-Year percentage change
plt.figure(figsize=(12, 6))
plt.plot(df_2['year'], df_2['Inflation Rate(%)'], marker='o', linestyle='-',color='magenta', markersize=5)
plt.xlabel('Year')
plt.ylabel('Percent')
plt.title('CPI for Fees and Admissions by Highest Education: Bachelors Degree Only')
plt.grid(True)

# Set x-axis to show only years
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y'))

# Set x-axis to show only years
plt.gca().xaxis.set_major_formatter(plt.matplotlib.ticker.FuncFormatter(lambda x, _: f'{int(x)}'))
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(integer=True))


# Annotate each point with the percentage change value
for i in range(len(df_2)):
    plt.text(df_2['year'].iloc[i], df_2['Inflation Rate(%)'].iloc[i]-5, 
             f"{df_2['Inflation Rate(%)'].iloc[i]:.2f}%", 
             ha='left', va='bottom', color='green', fontsize=10, rotation=10)
    
#Save the plot as an image in the specified folder
folder_name = "Images"
image_path = Path(folder_name, "Ed-CPI_bachelors.png")
plt.savefig(image_path)

plt.show()
```
