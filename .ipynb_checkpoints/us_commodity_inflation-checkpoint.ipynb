```python
# Save config information
#Item Code
#SA0	All items
#SAF	Food and beverages	0	T	3
# SETA	New and used motor vehicles	2	T	212
# SETA01	New vehicles	3	T	213
# SETA02	Used cars and trucks	3	T	218
# SETA03	Leased cars and trucks	3	T	219
# SETA04	Car and truck rental	3	T	220
# SETB	Motor fuel	2	T	221
# SETB01	Gasoline (all types)	3	T	222
# SETB02	Other motor fuels	3	T	226
# SETC	Motor vehicle parts and equipment	2	T	227
# SETC01	Tires	3	T	228
# SETC02	Vehicle accessories other than tires	3	T	229
# SETD	Motor vehicle maintenance and repair	2	T	232
# SETD01	Motor vehicle body work	3	T	233
# SETD02	Motor vehicle maintenance and servicing	3	T	234
# SETD03	Motor vehicle repair	3	T	235
# SETE	Motor vehicle insurance	2	T	236
# SETF	Motor vehicle fees	2	T	237
# SETF01	State motor vehicle registration and license fees	3	T	238
# SETF03	Parking and other fees	3	T	239
# SETG	Public transportation	1	T	242
# SETG01	Airline fares	2	T	243
# SETG02	Other intercity transportation	2	T	244
# SETG03	Intracity transportation	2	T	248

#Area Code
# 0000	U.S. city average
# 0100	Northeast
# 0110	New England
# 0120	Middle Atlantic
# 0200	Midwest
# 0230	East North Central
# 0240	West North Central
# 0300	South
# 0350	South Atlantic
# 0360	East South Central
# 0370	West South Central
# 0400	West
# 0480	Mountain
# 0490	Pacific

#Method to plot line graph 
def getFredObservationsBySeries(series_id, observation_start, observation_end):
    url = "https://api.stlouisfed.org/fred/series/observations"
    #geo_fred_url = "https://api.stlouisfed.org/geofred/series/observations"
    params = {
        "series_id": series_id,
        "observation_start": observation_start,
        "observation_end": observation_end,
        "api_key": fred_api_key,
        "file_type": "json"
    }
    # Fetch data
    response = requests.get(url, params=params)
    data = response.json()
    pp.pprint(data)
    # Convert to DataFrame
    observations = data['observations']
    observations_df = pd.DataFrame(observations)
    print(f"Observation DF Data Types {observations_df.describe}")
    #Convert the date column 
    observations_df['date'] = pd.to_datetime(observations_df['date'])
    observations_df.set_index('date', inplace=True)
    observations_df['value'] = observations_df['value'].astype(float)
    
    #calculate the inflation rate over 12 month period
    observations_df['Inflation Rate'] = observations_df['value'].pct_change(periods=12) * 100
    print(observations_df.head(10))
    # Drop rows with NaN values in column 'Inflation Rate'
    observations_df_cleaned = observations_df.dropna(subset=['Inflation Rate'])
    print(observations_df_cleaned.head(30))

    return observations_df_cleaned
    
    
#bls API call
# base_url = "https://api.bls.gov/publicAPI/v1/timeseries/data/"
# headers = {'Content-type': 'application/json'}
# payload = {
#     "seriesid": ['CUSR0000SA0','CUUR0000SA0L1E'],#,'APU0000701111'],
#     "startyear":"2014", 
#     "endyear":"2024",
#     "catalog":True|False,
#     "calculations":True|False,
#     "annualaverage":True|False,
#     "aspects":True|False,
#     "registrationkey":bls_api_key
# }

# # Create a dictionary to store DataFrames for multiple series
# dataframes = {}

# # Make the APi call uisng the JSON payload
# response = requests.post('https://api.bls.gov/publicAPI/v1/timeseries/data/', json=payload, headers=headers)
# #print(f"Response Status Code - {response.status_code}")
# response_data = response.json()
# #pp.pprint(response_data)
# series = response_data['Results']["series"]
# for entry in series:
#     # From each eries get the series_id
#     series_id = entry['catalog']['series_id']
#     print(series_id)
#     #Get the data element from the Series
#     data = entry['data']
#     df = pd.DataFrame(data)
#     #df.set_index('date', inplace=True)
#     dataframes[series_id] = df

# # Display the DataFrames
# for series_id, df in dataframes.items():
#     print(f"DataFrame for {series_id}:")
    
# cpi_df = pd.DataFrame(avg_cpi_data)
# print(cpi_df.head(30))


# url = "https://api.stlouisfed.org/fred/series/observations"
# #geo_fred_url = "https://api.stlouisfed.org/geofred/series/observations"
# params = {
#     "series_id": "CPIAUCSL",
#     "observation_start": "2013-08-04",
#     "observation_end": "2024-08-04",
#     "api_key": fred_api_key,
#     "file_type": "json"
# }
# # Fetch data
# response = requests.get(url, params=params)
# data = response.json()
# pp.pprint(data)
# # Convert to DataFrame
# observations = data['observations']
# observations_df = pd.DataFrame(observations)
# print(f"Observation DF Data Types {observations_df.describe}")
# #Convert the date column 
# observations_df['date'] = pd.to_datetime(observations_df['date'])
# observations_df.set_index('date', inplace=True)
# observations_df['value'] = observations_df['value'].astype(float)

# #calculate the inflation rate over 12 month period
# observations_df['Inflation Rate'] = observations_df['value'].pct_change(periods=12) * 100
# print(observations_df.head(10))
# # Drop rows with NaN values in column 'Inflation Rate'
# observations_df_cleaned = observations_df.dropna(subset=['Inflation Rate'])
# print(observations_df_cleaned.head(30))

# plt.figure(figsize=(10, 6))
# plt.plot(observations_df.index, observations_df['Inflation Rate'], label='Inflation Rate')
# plt.xlabel('Year')
# plt.ylabel('Inflation Rate (%)')
# plt.title('Consumer Price Index for All Urban Consumers: All Items in U.S. City Average (CPIAUCSL)')
# plt.legend()
# plt.show()

#Get the series for a given date range
#https://api.stlouisfed.org/fred/series?series_id=GNPCA&api_key=abcdefghijklmnopqrstuvwxyz123456&file_type=json
#Get Categories for a series
#https://api.stlouisfed.org/fred/series/categories?series_id=EXJPUS&api_key=abcdefghijklmnopqrstuvwxyz123456
#Regional data for a series
#https://api.stlouisfed.org/geofred/series/data?series_id=WIPCPI&api_key=abcdefghijklmnopqrstuvwxyz123456&file_type=json&date=2012-01-01
```

```python
#Use FRED API to fetch the Consumer Price Index for All Urban Consumers: All Items in U.S. City Average (Seasonally Adjusted)
cpi_urban_city_av_seasonal = 'CPIAUCSL'
#Use FRED API to fetch the Consumer Price Index for All Urban Consumers: All Items in U.S. City Average (Not Seasonally Adjusted)
cpi_urban_city_av_not_seasonal = 'CPIAUCNS'
#Use FRED API to fetch the Consumer Price Index for All Urban Consumers : All Items less Food and Energy U.S. City Average (Seasonally Adjusted)
cpi_urban_exclude_food_energy_sl = 'CPILFESL'
#Use FRED API to fetch the Consumer Price Index for All Urban Consumers : All Items less Food and Energy U.S. City Average (Not Seasonally Adjusted)
cpi_urban_exclude_food_energy_ns = 'CPILFENS'

ten_year_start_date = "2013-08-04"

covid_year_start_date = "2020-01-01"

russia_ukraine_war = "2022-02-01"

# Get the current date in format 'yyyy-mm-dd'
now = datetime.now()
now_iso_format = date(now.year,now.month,now.day).isoformat()
print("Current date and time:", now_iso_format)
```

```python
observations_au_sl_df = getFredObservationsBySeries(cpi_urban_city_av_seasonal,ten_year_start_date,now_iso_format)
observations_au_ns_df = getFredObservationsBySeries(cpi_urban_city_av_not_seasonal,ten_year_start_date,now_iso_format)
observations_lfe_sl_df = getFredObservationsBySeries(cpi_urban_exclude_food_energy_sl,ten_year_start_date,now_iso_format)
observations_lfe_ns_df = getFredObservationsBySeries(cpi_urban_exclude_food_energy_ns,ten_year_start_date,now_iso_format)
```

```python
# Plot the inflation rate that's CPI (Consumer Price Index for All Items in USA)
print(observations_au_sl_df.tail())
print(observations_au_ns_df.tail())
print(observations_lfe_sl_df.tail())
print(observations_lfe_ns_df.tail())

plt.figure(figsize=(20, 6))
plt.bar(observations_au_sl_df.index, observations_au_sl_df['Inflation Rate'], label='Inflation Rate (CPIAUCSL)',color='r')
plt.plot(observations_au_ns_df.index, observations_au_ns_df['Inflation Rate'], label='Inflation Rate (CPIAUCNS)',color='b')
plt.plot(observations_lfe_sl_df.index, observations_lfe_sl_df['Inflation Rate'], label='Inflation Rate (CPILFESL)',color='g')
plt.plot(observations_lfe_ns_df.index, observations_lfe_ns_df['Inflation Rate'], label='Inflation Rate (CPILFENS)',color='purple')
plt.plot(observations_lfe_ns_df.index, observations_lfe_ns_df['Inflation Rate'], label='Inflation Rate (CPIENGSL)',color='black')

plt.xlabel('Year')
plt.ylabel('Inflation Rate (%)')
plt.title('Consumer Price Index for All Urban Consumers: U.S. City Average')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame()
df = observations_au_sl_df
print(df.columns)
df = df.rename(columns={ 'Inflation Rate':'Cumulative Growth', 'value':'Value'})
# Sample data
# data = {
#     'Year': pd.date_range(start='1920', periods=100, freq='A'),
#     'Value': [i + (i % 10) * 2 for i in range(100)],
#     'Cumulative Growth': [i * 1.03 for i in range(100)]
# }
# df = pd.DataFrame(data)
print(df.info())
df['Year'] = df.index.year

print(df.head())

# Group by year
grouped = df.groupby(df['Year'])['Cumulative Growth'].sum()
g_df = pd.DataFrame(grouped)
print(g_df)

# Create the figure and axis
fig, ax1 = plt.subplots(figsize=(12, 6))

# Bar plot
ax1.bar(df['Year'], df['Value'], color='grey', label='Value')

# Line plot
ax2 = ax1.twinx()
ax2.plot(df['Year'], g_df['Cumulative Growth'], color='red', linestyle='dotted', label='Cumulative Growth')

# Annotations for specific events
events = {
    '1929-01-01': '1929 Stock Crash',
    '1933-01-01': '1933 Bank Failures & Output Collapse',
    '2008-01-01': '2008 Financial Crisis'
}
for date, event in events.items():
    ax1.axvline(pd.to_datetime(date), color='black', linestyle='--', linewidth=1)
    ax1.text(pd.to_datetime(date), max(df['Value']), event, rotation=90, verticalalignment='bottom')

# Add labels and title
ax1.set_xlabel('Year')
ax1.set_ylabel('Value')
ax2.set_ylabel('Cumulative Growth')
plt.title('Economic Data Over Time with Notable Events')

# Show the plot
plt.show()
```
